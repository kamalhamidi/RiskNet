| MÃ©trique | Formule | InterprÃ©tation |
|----------|---------|----------------|
| **Accuracy** | $\frac{TP+TN}{TP+TN+FP+FN}$ | Taux de prÃ©dictions correctes (biaisÃ© par majoritÃ©) |

| **Recall/SensibilitÃ©** | $\frac{TP}{TP+FN}$ | Proportion d'Ã©tudiants Ã  risque dÃ©tectÃ©s |
| **Specificity** | $\frac{TN}{TN+FP}$ | Proportion d'Ã©tudiants sÃ»rs correctement classÃ©s |
| **F1-Score** | $2 \cdot \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}$ | Harmonic mean (balance precision/recall) |
| **ROC-AUC** | $\int_0^1 \text{TPR}(FPR) \, dFPR$ | ProbabilitÃ© de ranking correct |

## 4. Gestion du Class Imbalance

### ProblÃ¨me IdentifiÃ©

DÃ©sÃ©quilibre de classe observÃ© : plus d'Ã©tudiants "non-risque" que "Ã -risque"

**ConsÃ©quences**:
- âŒ Biais du modÃ¨le vers la classe majoritÃ©
- âŒ MÃ©triques trompeuses (haute accuracy, faible recall)
- âŒ Non-dÃ©tection des cas positifs critiques

### StratÃ©gies ImplÃ©mentÃ©es

#### 1. Weighted Binary Cross-Entropy

$$L = -\sum_i [w_+ Â· y_i Â· \log(\hat{y}_i) + w_- Â· (1-y_i) Â· \log(1-\hat{y}_i)]$$

oÃ¹:
- $w_+ = \frac{N_{neg}}{N_{pos} + N_{neg}}$ (poids classe positive)
- $w_- = \frac{N_{pos}}{N_{pos} + N_{neg}}$ (poids classe nÃ©gative)

#### 2. Focal Loss (Optionnel)

## 3. Architecture GNN Hybride

### Justification ThÃ©orique

#### Combinaison Optimale: GCN + GAT

**1. GCN (Graph Convolutional Network)**

$$H^{(l+1)} = \sigma(\tilde{A} H^{(l)} W^{(l)})$$

oÃ¹ $\tilde{A} = D^{-1/2}A D^{-1/2}$ est la normalisation symÃ©trique.

- âœ… AgrÃ©gation lissÃ©e et stable
- âœ… ComplexitÃ© $O(|E|)$ efficace
- âœ… Capture les relations globales

**2. GAT (Graph Attention Network)**

$$\alpha_{ij} = \text{softmax}_j(\text{LeakyReLU}(a^T[Wh_i \| Wh_j]))$$

$$h_i' = \sigma\left(\sum_j \alpha_{ij} W h_j\right)$$

- âœ… Poids adaptatifs par arÃªte
- âœ… Identifie les relations critiques
- âœ… **InterpretabilitÃ©**: coefficients $\alpha_{ij}$
